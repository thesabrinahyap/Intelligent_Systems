```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs 
from sklearn.cluster import Birch
```

```python
data, clusters = make_blobs(n_samples = 1000, centers = 12, cluster_std = 0.50, random_state = 0)
data.shape
```

```python
plt.scatter(data[:, 0], data[:, 1],)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Synthetic Data Visualization')
plt.show()
```

```python
model = Birch(branching_factor = 50, n_clusters = None, threshold = 1.5)
```

```python
model.fit(data)
```

```python
pred = model.predict(data)
```

```python
plt.scatter(data[:, 0], data[:, 1], c = pred)
```

```python
from sklearn.cluster import KMeans

# Get the cluster centroids from the BIRCH model
cluster_centroids = model.subcluster_centers_

# Create a K-Means clustering model
kmeans_model = KMeans(n_clusters=3)  # Specify the desired number of global clusters

# Fit K-Means to the cluster centroids
kmeans_model.fit(cluster_centroids)

# Predict cluster labels for the cluster centroids
global_cluster_labels = kmeans_model.predict(cluster_centroids)

# Assign global cluster labels to data points based on the nearest centroid
global_cluster_assignments = global_cluster_labels[cluster_labels]

# Scatterplot to visualize the clusters after global clustering
plt.scatter(data[:, 0], data[:, 1], c=global_cluster_assignments, cmap='rainbow')
plt.title("BIRCH Clustering with Global Clustering (K-Means)")
plt.show()
```
